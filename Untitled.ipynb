```python
import seaborn as sns
import matplotlib.pyplot as plt

# Custom color palette (dark blue, sea green, red, yellow, orange)
custom_colors = ['#003366', '#2E8B57', '#FF0000', '#FFD700', '#FFA500', '#FF1493', '#4B0082']

plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x='Category',
    y='Profit',
    estimator=sum,
    hue='Category',
    palette=custom_colors,
    legend=False,
    errorbar=None
)

plt.title('Total Profit by Category', fontsize=16, color='#FF0000')
plt.xlabel('Category', fontsize=16)
plt.ylabel('Total Profit', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Top 10 Cities by Total Sales
top_cities_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12, 6))
sns. barplot(
    data=top_cities_sales,
    x='Sales',
    y='City',
    hue='City',          # 👈 Added hue
    palette='viridis',
    legend=False         # 👈 Hide legend
)
plt.title("Top 10 Cities by Total Sales", fontsize=14)
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
#Sales vs Profit by Category
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', palette='tab10', alpha=0.7)
plt.title("Sales vs Profit by Category", fontsize=14)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
import matplotlib.pyplot as plt

# Monthly Sales Trend
monthly_sales = df.resample('ME', on='Order Date')['Sales'].sum()  # 'ME' replaces 'M'

plt.figure(figsize=(14, 6))
monthly_sales.plot(color='darkseagreen', linewidth=4)  # 💚 Dark Sea Green line
plt.title("Monthly Sales Trend", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# --- Create a mock Payment Method column ---
np.random.seed(42)
df['Payment Method'] = np.random.choice(['Credit Card', 'Cash', 'Online Payment'], size=len(df))

# --- Plot Payment Method Distribution ---
plt.figure(figsize=(7, 7))
colors = sns.color_palette(['#1f77b4', '#2ca02c', '#ff7f0e'])  # blue, green, orange

df['Payment Method'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12, 'color': 'black'}
)

plt.title('Payment Method Distribution', fontsize=16, weight='bold', color='black')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
